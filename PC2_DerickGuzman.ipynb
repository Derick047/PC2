```python
import pandas as pd
data = "https://github.com/Derick047/PC2/raw/refs/heads/main/HDI.xlsx"
human_estad = pd.read_excel(data)
human_estad
```

**Quitando columnas y filas innecesarias**

```python
hum_estCopy= pd.read_excel(data, skiprows=3)
hum_estCopy.columns = hum_estCopy.iloc[0] #Asignando la primera fila como nombre de columna
hum_estCopy
```

```python
hum_estCopy = hum_estCopy.drop(index=[0,1,2,3]).reset_index(drop = True)
hum_estCopy
```

```python
hum_estCopy.columns.values[1] = "Country" #Rename
```

```python
hum_estCopy.columns
```

```python
hum_estCopy=hum_estCopy[['Country','Human Development Index (HDI) ','Life expectancy at birth','Expected years of schooling','Mean years of schooling','Gross national income (GNI) per capita']]
hum_estCopy #Nos quedamos unicamente con las columnas que nos interesan para el analisis
```

```python
hum_estCopy = hum_estCopy.replace("..", pd.NA)
```

```python
#Missing values?
hum_estCopy[hum_estCopy.isnull().any(axis=1)]
```

```python
hum_estCopy = hum_estCopy.dropna() #Eliminando las filas con Missing Values
hum_estCopy
```

```python
hum_estCopy.info() #Verificando el tipo de dato de cada columna
```

```python
pd.unique(hum_estCopy['Life expectancy at birth']
          [~hum_estCopy['Life expectancy at birth'].astype(str).str.contains(r'^\d+(?:\.\d+)?$', regex=True)])
```

```python
pd.unique(hum_estCopy['Human Development Index (HDI) ']
          [~hum_estCopy['Human Development Index (HDI) '].astype(str).str.contains(r'^\d+(?:\.\d+)?$', regex=True)])
```

```python
pd.unique(hum_estCopy['Expected years of schooling']
          [~hum_estCopy['Expected years of schooling'].astype(str).str.contains(r'^\d+(?:\.\d+)?$', regex=True)])
```

```python
pd.unique(hum_estCopy['Gross national income (GNI) per capita']
          [~hum_estCopy['Gross national income (GNI) per capita'].astype(str).str.contains(r'^\d+(?:\.\d+)?$', regex=True)])
```

**Como todos los valores tienen formato numérico válido, convertimos las columnas directamente a float**

```python
hum_estCopy['Human Development Index (HDI) ']=hum_estCopy.loc[: ,'Human Development Index (HDI) '].apply(lambda col:pd.to_numeric(col))
hum_estCopy['Expected years of schooling']=hum_estCopy.loc[: ,'Expected years of schooling'].apply(lambda col:pd.to_numeric(col))
hum_estCopy['Mean years of schooling']=hum_estCopy.loc[: ,'Mean years of schooling'].apply(lambda col:pd.to_numeric(col))
hum_estCopy['Gross national income (GNI) per capita']=hum_estCopy.loc[: ,'Gross national income (GNI) per capita'].apply(lambda col:pd.to_numeric(col))
```

```python
hum_estCopy.info()
```

**Agregamos una columna categorica para medir el nivel de HDI**


```python
hum_estCopy['HDI_Level']=pd.cut(hum_estCopy['Human Development Index (HDI) '],
                                bins=[0,0.25,0.5,0.75,1],
                                labels=['Muy bajo(0-0.25)', 'Bajo(0.25-0.5)', 'Medio(0.5-0.75)', 'Alto(0.75-1)'],
                                ordered=True)
hum_estCopy.head()
```
